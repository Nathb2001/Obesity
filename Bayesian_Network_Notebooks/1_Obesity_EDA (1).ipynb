```python
from ucimlrepo import fetch_ucirepo
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

# fetch dataset
estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition = fetch_ucirepo(id=544)

# data (as pandas dataframes)
X = estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.data.features
y = estimation_of_obesity_levels_based_on_eating_habits_and_physical_condition.data.targets
data=pd.concat([X,y],axis=1)
```

```python
data
```

```python
data.columns = ['Gender', 'Age','Height (CM)','Weight (KG)','Family History Overweight','Eat High Caloric Food Frequently','Eat Vegatables during each Meal','How Many Main Meals in a Day',
              'Eat Any Food between Meals','Smoke','Daily Water Intake','Monitor Daily Calories','How Often Physical Activity','How much time do you use technological devices',
              'Do you Drink Alcohol','Transport','Obesity Level']
```

```python
dataplot = sb.heatmap(data.corr(numeric_only=True), cmap="BuPu", annot=True)
```

```python
data['Obesity Level'] = data['Obesity Level'].replace({
    'Obesity_Type_I': 'Obese',
    'Obesity_Type_II': 'Obese',
    'Obesity_Type_III': 'Obese',
    'Overweight_Level_I': 'Overweight',
    'Overweight_Level_II': 'Overweight',
    'Normal_Weight': 'Normal',
    'Insufficient_Weight': 'Normal'
})
```

```python
data
```

```python
# @title Age

data['Age'].plot(kind='hist', bins=30, color='green', alpha=0.7, edgecolor='black', title='Age Distribution')

plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Age', fontsize=12)
plt.title('Age Distribution', fontsize=15)

# Show the plot
plt.tight_layout()  # Makes sure labels and title fit well
plt.show()
```

```python
# @title Weight

data['Weight (KG)'].plot(kind='hist', bins=30, color='navy', alpha=0.7, edgecolor='black')

plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Weight', fontsize=12)
plt.title('Weight', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
# @title Height

data['Height (CM)'].plot(kind='hist', bins=30, color='navy', alpha=0.7, edgecolor='black')

plt.grid(True, which='both', linestyle='--', linewidth=0.7)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Height', fontsize=12)
plt.title('Height', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
plt.scatter(data['Height (CM)'], data['Weight (KG)'], color='blue', alpha=0.7, edgecolor='black')

plt.title('Height vs. Weight Scatter Plot', fontsize=15)
plt.xlabel('Height (CM)', fontsize=12)
plt.ylabel('Weight (KG)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# @title Gender

data.groupby('Gender').size().plot(kind='barh',color=['pink', 'navy'],alpha=0.7, edgecolor='black')
plt.ylabel('Count',fontsize=12)
plt.xlabel('Gender',fontsize=12)
plt.title('Gender Distribution', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# @title Smoke

data.groupby('Smoke').size().plot(kind='barh', title='Smoking Distribution',color=['green', 'red'],alpha=0.7, edgecolor='black')
plt.ylabel('Count',fontsize=12)
plt.xlabel('Smoke',fontsize=12)
plt.tight_layout()
plt.show()
```

# Distributions between different Weight Groups

```python
df_obese = data[data['Obesity Level'] == 'Obese']
df_overweight = data[data['Obesity Level'] == 'Overweight']
df_normal = data[data['Obesity Level'] == 'Normal']
```

```python
fig, ax = plt.subplots()

ax.hist(df_obese['Weight (KG)'],color='red',edgecolor='black')
ax.hist(df_overweight['Weight (KG)'],color='yellow',edgecolor='black')
ax.hist(df_normal['Weight (KG)'],color='green',edgecolor='black')

ax.set_xlabel('Weight (KG)')

ax.set_ylabel('# of observations')

ax.set_title('Weight Distribution between different groups')

plt.show()
```

```python
fig, ax = plt.subplots()

# Add a boxplot for the "Height" column in the DataFrames
ax.boxplot([df_obese['Weight (KG)'],df_overweight['Weight (KG)'],df_normal['Weight (KG)']])

# Add x-axis tick labels:
ax.set_xticklabels(['Obese','Overweight','Normal'])

# Add a y-axis label
ax.set_ylabel('Weight (kg)')
ax.set_title('Boxplots of weight between weight groups')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(df_obese['Height (CM)'],color='red',edgecolor='black')
ax.hist(df_overweight['Height (CM)'],color='yellow',edgecolor='black')
ax.hist(df_normal['Height (CM)'],color='green',edgecolor='black')

ax.set_xlabel('Height (m)')

ax.set_ylabel('# of observations')

ax.set_title('Height Distribution between different groups')

plt.show()
```

```python
fig, ax = plt.subplots()

# Add a boxplot for the "Height" column in the DataFrames
ax.boxplot([df_obese['Height (CM)'],df_overweight['Height (CM)'],df_normal['Height (CM)']])

# Add x-axis tick labels:
ax.set_xticklabels(['Obese','Overweight','Normal'])

# Add a y-axis label
ax.set_ylabel('Height (cm)')
ax.set_title('Boxplots of height between weight groups')
plt.show()
```
